## Use the Obspy client to query a data center by name
### Disadvantages

1. You must know where is the data archived
2. You must check the acronym of the data center

```python
from obspy.clients.fdsn import Client
```

```python
from obspy import UTCDateTime
```

```python
# Create FDSN client
client = Client("GFZ")
```

```python
# Request waveforms
st = client.get_waveforms(network="GE", station="APE", location="*", channel="*", starttime=UTCDateTime(2016, 3, 3), endtime=UTCDateTime(2016, 3, 3, 0, 10, 0))
```

```python
print(st)
```

```python
st.plot()
```

```python
# Create FDSN client with debug turned on
client = Client("GFZ", debug=True)
```

```python
# Request waveforms
st = client.get_waveforms(network="GE", station="APE", location="*", channel="*", starttime=UTCDateTime(2016, 3, 3), endtime=UTCDateTime(2016, 3, 3, 0, 10, 0))
```
