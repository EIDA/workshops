## Download data using the API from FDSN Dataselect
### Disadvantages

1. You must know where is the data archived
2. You must know the domain name from the data center
3. You must build the URL and transform properly the data types. In particular the dates

```python
from obspy import read
```

```python
from obspy import UTCDateTime
```

```python
# Define domain name from data centre
domain = "http://geofon.gfz-potsdam.de"
path = "fdsnws/dataselect/1/query"
```

```python
# Define stream(s) and time window
net, sta, loc, cha = "GE.APE.*.*".split('.')
stime = UTCDateTime(2016, 3, 3)
etime = UTCDateTime(2016, 3, 3, 0, 10, 0)
```

```python
# Map parameters
params = ['net=%s' % net, 'sta=%s' % sta, 'loc=%s' % loc, 'cha=%s' % cha, 'start=%s' % stime, 'end=%s' % etime]
```

```python
# Show final URL
'%s/%s?%s' % (domain, path, '&'.join(params))
```

```python
# Read waveforms from the URL
st = read('%s/%s?%s' % (domain, path, '&'.join(params)))
```

```python
print(st)
```
